```python
import cv2
import numpy as np
```

```python
%pylab inline
```

```python
im = cv2.imread('ice_cream.png')
obj = cv2.imread('bee.png')
```

```python
mask = 255 * np.ones(obj.shape, obj.dtype)
```

```python
width, height, channels = im.shape
#center = (int(height/4), int(width/4))
center = (400,700)
```

```python
norm_clone = cv2.seamlessClone(obj, im, mask, center, cv2.NORMAL_CLONE)
```

```python
imshow(norm_clone[:,:,::-1])
```

```python
mix_clone = cv2.seamlessClone(obj, im, mask, center, cv2.MIXED_CLONE)
```

```python
imshow(mix_clone[:,:,::-1])
```
